```python
import tensorflow as tf
from tensorflow import keras

# Retrieve the dataset
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train_full.shape
X_train_full.dtype
```

```python
# Scale the train set
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
# Define the name of the classes 
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
# Example
class_names[y_train[0]]
# plt.imshow(X_train[0], cmap="binary")
# plt.axis("off")
# plt.show()
```

```python
# Build our ANN
# Classification MLP with 2 hidden layers
model = keras.models.Sequential([    
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

model.summary()
```

```python
# Get the model's list of layers
model.layers
# Retrieve the name of the hidden layer 1
model.layers[1].name
model.get_layer('dense_9').name
# Retrieve the weights and the biases of the first hidden layer
weights, biases = model.layers[1].get_weights()
biases.shape
weights.shape
```

```python
# Specify the loss function 
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])

```

```python
# train the model
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_valid, y_valid))
```

```python
# Plot the history of the training
import pandas as pd
# pd.DataFrame(history.history).plot(figsize=(8,5))
# plt.grid(True)
# plt.ylim([0, 1])
# plt.show()

```

```python
# Evaluate the model
model.evaluate(X_test, y_test)

```

```python
# Using the model to make predictions
X_new = X_test[266:269]
y_proba = model.predict(X_new)
y_proba.round(2)

```
